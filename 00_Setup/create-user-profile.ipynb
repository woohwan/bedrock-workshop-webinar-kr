## 선수 사항
SageMakerFullAccess 에는 아래와 같은 권한이 기본적으로 없습니다. 노트북을 실행하는 역할에 아래 권한을 추가 해주세요.
```
{
	"Version": "2012-10-17",
	"Statement": [
		{
			"Sid": "VisualEditor0",
			"Effect": "Allow",
			"Action": "sagemaker:CreateUserProfile",
			"Resource": "*"
		}
	]
}
```
